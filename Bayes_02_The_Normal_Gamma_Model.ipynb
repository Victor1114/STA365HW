### 3. What distribution do you think this $\theta$ has?

$\large
\begin{align*}
\prod_{i=1}^n\frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{1}{2\sigma^2}(x_i-\theta)^2} &={} (2\pi\sigma^2)^{-\frac{n}{2}} e^{-\frac{1}{2\sigma^2} \sum_{i=1}^n (x_i-\theta)^2}\\

\textrm{So} & \;\;\; \theta \sim N\left(\bar x, \frac{\sigma^2}{n}]\right)
\end{align*}

$\begin{align*}
p\left(x|\theta,\tau=\frac{1}{\sigma^2}\right) = {} & \prod_{i=1}^n \sqrt{\frac{\tau}{2\pi}} e^{-\frac{\tau}{2}(x_i-\theta)^2}\\

p(\tau|\theta, x) = {} & \text{Gamma}\left(\alpha = \frac{n}{2}+1, \textrm{rate} = \frac{1}{2} \sum_{i=1}^n (x_i-\theta)^2 \right)\\

  \propto {} & e^{-\frac{n \tau}{2} (\theta - \bar x)^2} \\
p(\theta| \sigma^2 = \tau^{-1}, x) = {} & N\left(E[\theta] = \bar x, Var(\theta) = \frac{\sigma^2}{n}\right)
\end{align*}$


```python
import numpy as np
from scipy.stats import norm, gamma
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# loc is mean and scale is standard deviation
n = 30; x = stats.norm(loc=0, scale=1).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples
theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization
for c in range(C):
    for g in range(1,G):
        # https://en.wikipedia.org/wiki/Gamma_distribution
        tau[c,g] = None
        theta[c,g] = None
```

```python
# loc is mean and scale is standard deviation
n = 30; x = stats.norm(loc=0, scale=1).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples
theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization
for c in range(C):
    for g in range(1,G):
        # https://en.wikipedia.org/wiki/Gamma_distribution
        shape_tau = n / 2
        rate_tau = 0.5 * np.sum((x - theta[c, g - 1]) ** 2)
        tau[c, g] = gamma(a=shape_tau, scale=1 / rate_tau).rvs()

        mean_theta = np.mean(x)
        variance_theta = 1 / (n * tau[c, g])
        theta[c, g] = stats.norm(loc=mean_theta, scale=np.sqrt(variance_theta)).rvs()
```

## Q2

### 1. Derive \( p($\tau | \theta, \mathbf{x})$)

Using Bayes' theorem:
$$
p(\tau | \theta, \mathbf{x}) \propto p(\mathbf{x} | \theta, \tau) p(\tau)
$$

Since $(p(\mathbf{x} | \theta, \tau))$ is a normal distribution, the joint distribution becomes:
$$
p(\tau | \theta, \mathbf{x}) \propto \tau^{n/2} e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i - \theta)^2} \cdot \tau^{\alpha/2 - 1} e^{-\frac{\lambda}{2} \tau}
$$

Simplify:
$$
p(\tau | \theta, \mathbf{x}) \propto \tau^{\frac{n+\alpha}{2} - 1} e^{-\frac{\tau}{2} \left( \sum_{i=1}^n (x_i - \theta)^2 + \lambda \right)}
$$

This is a Gamma distribution:
$$
\tau | \theta, \mathbf{x} \sim \text{Gamma}\left( \frac{n + \alpha}{2}, \frac{\sum_{i=1}^n (x_i - \theta)^2 + \lambda}{2} \right)
$$

### 2. Derive \( p($\theta | \tau, \mathbf{x}$) \)

Using Bayes' theorem:
$$
p(\theta | \tau, \mathbf{x}) \propto p(\mathbf{x} | \theta, \tau) p(\theta | \tau)
$$

The likelihood is:
$$
p(\mathbf{x} | \theta, \tau) \propto e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i - \theta)^2}
$$

The prior for \($\theta$) is:
$$
p(\theta | \tau) \propto e^{-\frac{1}{2\tau_0^2} (\theta - \theta_0)^2}
$$

The joint distribution becomes:
$$
p(\theta | \tau, \mathbf{x}) \propto e^{-\frac{\tau}{2} \sum_{i=1}^n (x_i - \theta)^2} \cdot e^{-\frac{1}{2\tau_0^2} (\theta - \theta_0)^2}
$$

Simplify:
$$
p(\theta | \tau, \mathbf{x}) \propto e^{-\frac{1}{2} \left( \tau n (\theta - \bar{x})^2 + \frac{1}{\tau_0^2} (\theta - \theta_0)^2 \right)}
$$

This is a Normal distribution:
$$
\theta | \tau, \mathbf{x} \sim N\left( \mu, \sigma^2 \right)
$$

Where:
$$
\mu = \frac{\tau n \bar{x} + \frac{\theta_0}{\tau_0^2}}{\tau n + \frac{1}{\tau_0^2}}, \quad \sigma^2 = \frac{1}{\tau n + \frac{1}{\tau_0^2}}
$$

## Step 2: Role of Hyperparameters $( \alpha) and ( \lambda)$

- **($\alpha$):** Controls the shape of the prior for ($\tau$). Larger ($\alpha$) gives more weight to the prior, contributing more confidence.
- **($\lambda$):** Controls the scale of the prior for ($\tau$). Larger ($\lambda$) makes the prior more concentrated.

The prior can be seen as contributing a "virtual dataset," and by adjusting the hyperparameters \($\alpha$) and \($\lambda$), the influence of the prior relative to the data can be controlled.


```python
# Updated Gibbs Sampler
n = 30; x = stats.norm(loc=0, scale=1).rvs(size=n)
C,G = 2,10000  # chains, Gibbs samples
theta,tau = np.zeros([C,G]),np.zeros([C,G])
theta[:,0] = 1000  # initialization
alpha, lambda_ = 2, 1  # hyperparameters for gamma prior
theta_0, tau_0 = 0, 1

for c in range(C):
    for g in range(1,G):
        # https://en.wikipedia.org/wiki/Gamma_distribution
        shape_tau = (n + alpha) / 2
        rate_tau = 0.5 * (np.sum((x - theta[c, g - 1]) ** 2) + lambda_)
        tau[c, g] = gamma(a=shape_tau, scale=1 / rate_tau).rvs()

        mean_theta = (np.mean(x) * tau[c, g] * n + theta_0 / tau_0**2) / (tau[c, g] * n + 1 / tau_0**2)
        variance_theta = 1 / (tau[c, g] * n + 1 / tau_0**2)
        theta[c, g] = stats.norm(loc=mean_theta, scale=np.sqrt(variance_theta)).rvs()
```

```python
fig,ax = plt.subplots(1, 4, figsize=(14,3)); ax[0].set_title("$\\theta$ Burn-In"); ax[1].set_title("$\\tau$ Burn-In"); ax[2].set_title("$\\theta$ Converged Markov Chain\nStationary Distribution Samples"); ax[3].set_title("$\\tau$ Converged Markov Chain\nStationary Distribution Samples")
burn = 20
demo = 120
for c in range(C):
    ax[0].plot(theta[c,:burn], label="$\\theta$ Chain "+str(c))
    ax[1].plot(tau[c,:burn], label="$\\tau$ Chain "+str(c))
    ax[2].plot(np.arange(burn, demo, dtype=int), theta[c,burn:demo], label="$\\theta$ Chain "+str(c))
    ax[3].plot(np.arange(burn, demo, dtype=int), tau[c,burn:demo], label="$\\tau$ Chain "+str(c))
ax[0].legend(); ax[1].legend(); ax[2].legend(); ax[3].legend();
```

```python
import numpy as np
from scipy.stats import norm, gamma
import matplotlib.pyplot as plt

# Set hyperparameters
alpha_values = [1, 5, 10]  # Different values for alpha
lambda_values = [1, 10, 50]  # Different values for lambda
theta_0 = 0  # Mean of prior for theta
tau_0_values = [1, 5, 10]  # Different values for tau_0

# Data setup
n = 30
x = norm(loc=0, scale=1).rvs(size=n)  # Simulate observed data
x_bar = np.mean(x)

# Posterior parameters for tau
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
for i, alpha in enumerate(alpha_values):
    for j, lambd in enumerate(lambda_values):
        # Gamma posterior for tau
        shape_tau = (n + alpha) / 2
        rate_tau = 0.5 * (np.sum((x - x_bar) ** 2) + lambd)
        tau_posterior = gamma(a=shape_tau, scale=1 / rate_tau)

        # Generate samples and plot
        tau_samples = tau_posterior.rvs(1000)
        axes[i, j].hist(tau_samples, bins=30, density=True, alpha=0.7, color='blue')
        axes[i, j].set_title(rf"$\alpha={alpha}, \lambda={lambd}$")
        axes[i, j].set_xlabel(r"$\tau$")
        axes[i, j].set_ylabel("Density")

plt.tight_layout()
plt.suptitle("Posterior Distributions of $\\tau$ with Different $\\alpha$ and $\\lambda$", y=1.02, fontsize=16)
plt.show()

# Posterior parameters for theta
fig, axes = plt.subplots(1, len(tau_0_values), figsize=(15, 5))
for k, tau_0 in enumerate(tau_0_values):
    variance_theta = 1 / (n + 1 / tau_0**2)
    mean_theta = (x_bar * n + theta_0 / tau_0**2) * variance_theta
    theta_posterior = norm(loc=mean_theta, scale=np.sqrt(variance_theta))

    # Generate samples and plot
    theta_samples = theta_posterior.rvs(1000)
    axes[k].hist(theta_samples, bins=30, density=True, alpha=0.7, color='green')
    axes[k].set_title(rf"$\tau_0={tau_0}$")
    axes[k].set_xlabel(r"$\theta$")
    axes[k].set_ylabel("Density")

plt.tight_layout()
plt.suptitle("Posterior Distributions of $\\theta$ with Different $\\tau_0$", y=1.02, fontsize=16)
plt.show()
```

### How Hyperparameters Affect the Posterior:
- **\($\alpha$):** Controls the shape of the Gamma prior for \($\tau$). Larger \($\alpha$) provides stronger prior influence.
- **\($\lambda$):** Scales the Gamma prior for \($\tau$). Larger \($\lambda$) leads to a more concentrated prior.
- **\($\theta_0 $):** The mean of the Normal prior for \($\theta$). Smaller deviations from \($\theta_0 $) are favored when \($\tau_0^2$) is small.
- **\($\tau_0^2 $):** The variance of the Normal prior for \($\theta$). Smaller \($\tau_0^2 $) increases the weight of \($\theta_0$) in the posterior.

### Assumed Model

Suppose we have the following model setup:  
$$
x_1, x_2, \ldots, x_K \sim \text{Multinomial}(n, \boldsymbol{\theta})
$$  
where $\boldsymbol{\theta} = (\theta_1, \theta_2, \ldots, \theta_K)$ satisfies:  
$$
\sum_{i=1}^K \theta_i = 1, \quad \theta_i > 0.
$$  

The prior distribution for $\boldsymbol{\theta}$ is:  
$$
\boldsymbol{\theta} \sim \text{Dirichlet}(\alpha_1, \alpha_2, \ldots, \alpha_K)
$$  

---

### Joint Distribution

The joint likelihood distribution of $x_1, x_2, \ldots, x_K$ given $\boldsymbol{\theta}$ is:  
$$
p(\mathbf{x} | \boldsymbol{\theta}) = \frac{n!}{x_1! x_2! \cdots x_K!} \prod_{i=1}^K \theta_i^{x_i}.
$$  

The prior distribution is:  
$$
p(\boldsymbol{\theta}) = \frac{\Gamma(\sum_{i=1}^K \alpha_i)}{\prod_{i=1}^K \Gamma(\alpha_i)} \prod_{i=1}^K \theta_i^{\alpha_i - 1}.
$$  

Combining the two, the joint distribution of $\boldsymbol{\theta}$ and $\mathbf{x}$ is:  
$$
p(\mathbf{x}, \boldsymbol{\theta}) = p(\mathbf{x} | \boldsymbol{\theta}) p(\boldsymbol{\theta})
$$  
$$
= \frac{n!}{x_1! x_2! \cdots x_K!} \prod_{i=1}^K \theta_i^{x_i} \cdot \frac{\Gamma(\sum_{i=1}^K \alpha_i)}{\prod_{i=1}^K \Gamma(\alpha_i)} \prod_{i=1}^K \theta_i^{\alpha_i - 1}.
$$  

---

### Posterior Distribution

The posterior distribution is given by Bayes' theorem:  
$$
p(\boldsymbol{\theta} | \mathbf{x}) = \frac{p(\mathbf{x} | \boldsymbol{\theta}) p(\boldsymbol{\theta})}{p(\mathbf{x})}.
$$  
Ignoring the normalization constant $p(\mathbf{x})$, we have:  
$$
p(\boldsymbol{\theta} | \mathbf{x}) \propto p(\mathbf{x} | \boldsymbol{\theta}) p(\boldsymbol{\theta}).
$$  

Substituting the likelihood and prior:  
$$
p(\boldsymbol{\theta} | \mathbf{x}) \propto \left( \prod_{i=1}^K \theta_i^{x_i} \right) \cdot \left( \prod_{i=1}^K \theta_i^{\alpha_i - 1} \right).
$$  

Combining the exponents:  
$$
p(\boldsymbol{\theta} | \mathbf{x}) \propto \prod_{i=1}^K \theta_i^{x_i + \alpha_i - 1}.
$$  

Comparing with the standard form of the Dirichlet distribution:  
$$
\text{Dirichlet}(\boldsymbol{\theta} | \boldsymbol{\beta}) = \frac{\Gamma(\sum_{i=1}^K \beta_i)}{\prod_{i=1}^K \Gamma(\beta_i)} \prod_{i=1}^K \theta_i^{\beta_i - 1},
$$  
we see that the posterior is still a Dirichlet distribution, with updated parameters:  
$$
\alpha_i^{\text{posterior}} = \alpha_i^{\text{prior}} + x_i.
$$  

---

### Conclusion

The posterior distribution for the Multinomial-Dirichlet model remains a Dirichlet distribution, with parameters updated as:  
$$
\boldsymbol{\alpha}^{\text{posterior}} = \boldsymbol{\alpha}^{\text{prior}} + \mathbf{x}.
$$  

This demonstrates the **conjugacy** property of the Dirichlet distribution as a prior for the Multinomial likelihood.
